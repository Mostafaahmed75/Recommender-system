```python
import pandas as pd 
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
my_df = pd.read_csv('C:\\Users\\HP\\Downloads\\archive\\Datafiniti_Amazon_Consumer_Reviews_of_Amazon_Products.csv', usecols=['id','primaryCategories','reviews.rating','name','reviews.username'])
```

```python
my_df.head()
```

```python
ratings = pd.DataFrame(my_df.groupby('name')['reviews.rating'].mean())
ratings.head()
```

```python
ratings['number_of_ratings'] = my_df.groupby('name')['reviews.rating'].count()
ratings.head(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.jointplot(x='reviews.rating', y='number_of_ratings', data=ratings)
```

```python
items_matrix_UII = my_df.pivot_table(index='reviews.username', columns='name', values='reviews.rating')
items_matrix_UII.head()
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
Echo_user_rating = items_matrix_UII['Amazon Echo Show Alexa-enabled Bluetooth Speaker with 7" Screen']
```

```python
similar_to_Echo=abs(items_matrix_UII.corrwith(Echo_user_rating))
```

```python
similar_to_Echo.head(10)
```

```python
corr_Echo = pd.DataFrame(similar_to_Echo, columns=['Correlation'])
corr_Echo.dropna(inplace=True)
corr_Echo.head()
```

```python
corr_Echo = corr_Echo.join(ratings['number_of_ratings'])

corr_Echo.head()
```

```python
corr_Echo[corr_Echo['number_of_ratings'] > 100].sort_values(by='Correlation', ascending=False).head(7)
```
